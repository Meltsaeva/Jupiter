2 Выберите дом клиенту.
Клиент хочет дом
- либо с очень красивым видом(параметр 'view' - больше 3) из окна, либо с видом на набережную(waterfront)
- и в очень хорошем состоянии(condition > 4) и год постройки не меньше 1980 года.
- В какой ценовом диапазоне будут дома?

```python
df[(df['view'] >3) | (df['waterfront'] == 1) & (df['condition'] > 4) & (df['yr_built'] > 1980)]
```

```python
df[(df['view'] >3) | (df['waterfront'] == 1) & (df['condition'] > 4) & (df['yr_built'] > 1980)].describe()
```

```python
df[(df['view'] >3) | (df['waterfront'] == 1) & (df['condition'] > 4) & (df['yr_built'] > 1980)]['price'].min()
```

```python
df[(df['view'] >3) | (df['waterfront'] == 1) & (df['condition'] > 4) & (df['yr_built'] > 1980)]['price'].max()
```

3 Выберите дом клиенту. Клиент хочет дом
- без подвала
- с двумя этажами
- стоимостью до 150000.
- Какая оценка по состоянию у таких домов в среднем?

```python
df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] < 150000)]['condition'].mean()
```
