1. Выберите ноутбук клиенту. 
Клиент хочет подобрать ноутбук с 8GB или 16GB ОЗУ на Windows 10 в стоимости до 500 евро, сколько у него вариантов?



```python
import pandas as pd
```

```python
df = pd.read_csv('laptop_price.csv', encoding= 'cp1251')
```

```python
df[(df['OpSys'] == 'Windows 10') & ((df['Ram'] == '16GB') | (df['Ram'] == '8GB')) & (df['Price_euros'] <= 500)]
```

```python
df[(df['OpSys'] == 'Windows 10') & ((df['Ram'] == '16GB') | (df['Ram'] == '8GB')) & (df['Price_euros'] <= 500)].shape[0]
```

2. Выберите ноутбук клиенту.
